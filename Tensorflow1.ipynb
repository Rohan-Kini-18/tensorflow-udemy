**Positional Min or max tensor**

```python
tf.random.set_seed(42)
a=tf.random.uniform(shape=[50])
a
```

1) Positional Max

```python
# finding positional maximum
tf.argmax(a)
```

```python
# checking largest value by above index value
a[tf.argmax(a)]
```

```python
# getting value again
tf.reduce_max(a)
```

```python
# checking error and equality
assert a[tf.argmax(a)]==tf.reduce_max(a)
```

```python
# comparring max values
a[tf.argmax(a)]==tf.reduce_max(a)
```

2) Positional min

```python
# finding positional minimum
tf.argmin(a)
```

```python
# checking largest value by above index value
a[tf.argmin(a)]
```

```python
# getting value again
tf.reduce_min(a)
```

```python
# checking error and equality
assert a[tf.argmin(a)]==tf.reduce_min(a)
```

```python
# comparring min values
a[tf.argmin(a)]==tf.reduce_min(a)
```
